Messy Data Clean Up Begins

```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the data into a dataframe
df = pd.read_csv('C:/Users/Rafael/Documents/GitHub/messy_dataset/messy_dataset.csv')
df_original = df.copy() # Keeps a copy of the initial dataset in memory
df.head()
```

Cleaning the Name Columns

```python
# Checking the total number of each unique names available in the dataset. This shows that the names need to be standardized
df['FirstName'].value_counts()
```

```python
df['LastName'].value_counts()
```

```python
# df['Name'] = df['Name'].str.strip() # Removing all the spaces in around the name

df['FirstName'] = df['FirstName'].str.strip('/_123. ') # This removes all the outlined items to be striped off of the column
df['LastName'] = df['LastName'].str.strip('/_123. ')

df['FirstName'] = df['FirstName'].str.lower().str.title() # Converts all the texts to Camel Cases
df['LastName'] = df['LastName'].str.lower().str.title()   # Converts all the texts to Camel Cases
```

```python
df['FirstName'].value_counts()
```

```python
df['LastName'].value_counts()
```

```python
df[['FirstName', 'LastName']].value_counts()
```

```python
# Standardizing the names by replacing the incorrect ones with the correct ones

df['FirstName'] = df['FirstName'].replace({ \
                                        'john': 'John',
                                        'Jnae': 'Jane',
                                        'KHALID': 'Khalid',
                                        'Francyn': 'Francyn'
                                    })

df['LastName'] = df['LastName'].replace({ \
                                        'Deo': 'Doe'
                                    })
```

```python
# Filling the Null names with empty values
cols_to_fill = ['FirstName', 'LastName']
df[cols_to_fill] = df[cols_to_fill].fillna('')

df.head(10)
```

```python
df['FirstName'].value_counts()
```

```python
df['LastName'].value_counts()
```

Cleaning the Age Column

```python
# Checking the unique ages in the dataframe
df['Age'].value_counts()
```

```python
df['Age'] = df['Age'].str.strip() # Removing all unwanted spaces around the ages (if there are any)
df['Age'] = df['Age'].replace('forty', 40) # Replacing the word forty with the number 40
```

```python
df['Age'].value_counts()
```

```python
# Convert the age column to integer. Int64 converts the column to nullable integers which keeps NaN and <NA>
df['Age'] = pd.to_numeric(df['Age']).astype('Int64')
```

```python
# Filters the dataframe with only the valid ages and stores it back in the dataframe
df = df[(df['Age'] > 0) & (df['Age'] < 120)]
```

```python
df.head(10)
```

Cleaning the Salary Column

```python
df['Salary'].value_counts()
```

```python
df.isnull().sum() # Salary had 142 Null values here
```

```python
df['Salary'] = df['Salary'].str.strip('?/, ') # Removing all unwanted characters around the values
# df['Salary'] = df['Salary'].replace({'?': '', ',': '', '/': '',' ': ''})
df['Salary'] = df['Salary'].str.replace(',', '') # Removing the comma from the salary values that are stored as strings
df['Salary'] = pd.to_numeric(df['Salary']).astype('Int64') # Converting the salary column to integer which also converts all invalid values to NULL or <NA>
df['Salary'].value_counts()
```

```python
df.isnull().sum() # Salary column now has 235 null values cos of this conversion "df['Salary'] = pd.to_numeric(df['Salary']).astype('Int64')"
```

```python
# Filling all the Missing values with the floor of the average salaries
df['Salary'] = df['Salary'].fillna((df['Salary'].mean() // 1000) * 1000)
```

Cleaning the Date Column

```python
# df['JoinDate']
# pd.to_datetime(df['JoinDate'], errors = 'coerce', infer_datetime_format= True)
# pd.to_datetime(df['JoinDate'].astype(str), errors='coerce')

# pd.to_datetime(
#     df['JoinDate'].astype(str).str.replace('[^0-9/]', '', regex=True),  # remove unwanted chars
#     format=None,  # let pandas infer format
#     errors='coerce',  # invalid ones become NaT
# )

df['JoinDate'] = pd.to_datetime(df['JoinDate'], format= 'mixed') # Worked effectively because of the new pandas format='mixed' that handles all mixed date formats properly
df['JoinDate'] = df['JoinDate'].ffill() # (Forward Fill) Fills the NULL values with the value of the previous record (bfill() for backward fill)
```

Cleaning The Departments Column

```python
df['Department'].value_counts()
```

```python
df['Department'] = df['Department'].str.strip('. ')
df['Department'] = df['Department'].str.lower().str.title() # First make everything lower case, before converting them to Camel Case
df['Department'] = df['Department'].replace(
    {
        'Eng': 'Engineering',
        'FIN': 'Finance',
        'Fin': 'Finance'
    })

df['Department'].value_counts()
```

```python
df['Department'].isnull().sum()
```

```python
df['Department'] = df['Department'].fillna('Unknown')
```

```python
df.isnull().sum()
```

```python
df.isnull().sum()
```

Cleaning The Notes Column

```python
df['Notes'].value_counts()
```

```python
df['Notes'] = df['Notes'].astype('str')
df['Notes'] = df['Notes'].str.strip('???, ')

df['Notes'] = df['Notes'].replace('', 'No Review')
df['Notes'] = df['Notes'].replace('No review', 'No Review')
df['Notes'] = df['Notes'].replace('Good', 'Good Employee')
df['Notes'] = df['Notes'].replace('Nan', 'No Review')


df['Notes'] = df['Notes'].str.lower().str.title()
df['Notes'] = df['Notes'].fillna('No Review')

df['Notes'].value_counts()
```
